```python
import osmnx as ox
from IPython.display import Image
%matplotlib inline
ox.config(log_console=True, use_cache=True)
ox.__version__
```

```python
# configure the inline image display
img_folder = 'images'
extension = 'png'
size = 240
dpi = 40
```

### San Francisco

```python
place = 'sf'
point = (37.793897, -122.402189)
fig, ax = ox.plot_figure_ground(point=point, dpi=dpi)
```

### Barcelona

```python
place = 'barcelona-gothic'
point = (41.382850, 2.174192)
fig, ax = ox.plot_figure_ground(point=point, network_type='all', default_width=3.3, dpi=dpi)
```

### Paris

```python
place = 'paris'
point = (48.873702, 2.294136)
fig, ax = ox.plot_figure_ground(point=point, network_type='drive', default_width=3.3, dpi=dpi)
```
